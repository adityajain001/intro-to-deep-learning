## 2. Variables, Types & Indentation

Python is **dynamically typed** so you don't declare a type, it's inferred at runtime. The four basic types you'll use constantly: `int`, `float`, `str`, `bool`.

Python also uses **indentation**, not curly braces, to define code blocks. Indentation is not a style choice here, it's syntax. Get it wrong and your code either errors out or silently does the wrong thing.


```python
x = 5          # int
y = 5.0        # float
name = "Ada"   # str
is_ready = True  # bool

print(type(x), type(y), type(name), type(is_ready))

# Indentation defines the block:
if is_ready:
    print("Ready to go")
else:
    print("Not ready yet")
```

## 3. Operators

Standard arithmetic, plus two Python-specific ones worth knowing cold: **floor division** (`//`) and **exponentiation** (`**`). You'll use both constantly once you're computing output shapes for conv layers.


```python
a, b = 10, 3

print("a + b  =", a + b)
print("a - b  =", a - b)
print("a * b  =", a * b)
print("a / b  =", a / b)    # true division -> float
print("a // b =", a // b)   # floor division -> int (rounds down)
print("a % b  =", a % b)    # modulo (remainder)
print("a ** b =", a ** b)   # exponent
```

## 4. Booleans & Logical Operators

`and`, `or`, `not` are used everywhere in conditionals, loops, and filtering logic.


```python
A = True
B = False

print("A and B:", A and B)   # False - both must be True
print("A or B :", A or B)    # True  - only one needs to be True
print("not A  :", not A)     # False - flips the value
```

## 5. Strings & f-strings

Strings are **immutable**. Methods like `.lower()`, `.strip()`, `.split()` return a *new* string rather than modifying in place. `.strip()` in particular will save you constantly when cleaning messy data.

**f-strings** (`f"..."`) let you embed variables and expressions directly inside a string, including format specifiers like decimal precision.


```python
s = "  Hello, Deep Learning!  "
print(s.lower())
print(s.strip())
print(s.strip().split(","))

name = "John"
age = 30
print(f"Hello {name}, you are {age} years old.")

import math
print(f"Pi to 2 decimals: {math.pi:.2f}")
print(f"Pi to 5 decimals: {math.pi:.5f}")
```

## 6. Data Structures: Lists, Tuples, Dicts, Sets

| Structure | Ordered | Mutable | Duplicates | Syntax |
|---|---|---|---|---|
| `list` | Yes | Yes | Yes | `[1, 2, 3]` |
| `tuple` | Yes | **No** | Yes | `(1, 2, 3)` |
| `dict` | Yes (3.7+) | Yes | Unique keys | `{"a": 1}` |
| `set` | No | Yes | **No** | `{1, 2, 3}` |


```python
# Lists are mutable and ordered
nums = [1, 2, 3]
nums.append(4)
print("list:", nums)

# Tuples are immutable and ordered. Good for fixed records like coordinates.
point = (3, 4)
try:
    point[0] = 5
except TypeError as e:
    print("Tuple error:", e)

# Dictionaries are key/value pairs
student = {"name": "Alice", "score": 95}
print("dict:", student, "| name:", student["name"])

# Sets are unique, unordered
s = {1, 4, 2, 3, 2}
print("set:", s)
```

## 7. Comprehensions

A compact way to build a list/dict/set from an iterable in a single line. You'll see these constantly in homework starter code, so it's worth being comfortable reading and writing them.


```python
nums = [1, 2, 3, 4, 5]

squares = [n ** 2 for n in nums]
odd_squares = [n ** 2 for n in nums if n % 2 == 1]
num_to_cube = {n: n ** 3 for n in nums}

print("squares:    ", squares)
print("odd_squares:", odd_squares)
print("num_to_cube:", num_to_cube)
```

## 8. Conditionals

`if` / `elif` / `else`. Only the first `True` branch runs; the rest are skipped.


```python
number = -5

if number > 0:
    print("positive number")
elif number == 0:
    print("zero")
else:
    print("negative number")
```

## 9. Loops

Use a `for` loop when you know what you're iterating over; use `while` when you're looping until some condition changes. `range(5)` starts at `0` and stops *before* `5`, that's the single most common off-by-one mistake in this class, so get used to it now.

`enumerate` and `zip` are two built-ins you'll reach for a lot: `enumerate` gives you the index alongside the value, `zip` lets you iterate over two sequences in parallel.

**Heads up:** in deep learning, your data is often huge. Looping in pure Python over every element is slow. Whenever you can, you'll want **vectorized** NumPy/PyTorch operations instead of explicit loops  (more on this in the NumPy section below).


```python
for i in range(5):
    print(i, end=" ")
print()  # 0 1 2 3 4 --> never reaches 5

x = 3
while x > 0:
    print(x, end=" ")
    x -= 1
print()

names = ["alice", "bob", "carol"]
for i, n in enumerate(names):
    print(i, n)

scores = [90, 85, 70]
for n, sc in zip(names, scores):
    print(n, "->", sc)
```

## 10. Functions, *args / **kwargs, and Type Hints

`def` defines a function. `*args` collects extra positional arguments into a tuple; `**kwargs` collects extra keyword arguments into a dict. This is what lets a function accept a variable number of inputs, very common in framework code.

Type hints (`def f(x: int) -> int:`) are optional and **not enforced at runtime**, they're purely for readability and for your IDE/type checker. Get used to reading them; most homework starter code uses them.


```python
def greet(name):
    print(f"Hello {name}")

greet("Bob")

def describe(student_id, *args, **kwargs):
    print("ID:", student_id)
    print("positional extras:", args)
    print("keyword extras:", kwargs)

describe(101, "extra_arg", major="ML", year=2)

def add(a: int, b: int) -> int:
    return a + b

print(add(2, 3))
```

## 11. Mutability & Pass-by-Reference

Variables in Python are passed **by reference**. For mutable objects like lists, a function can modify the original object in place (changes persist outside the function). This trips people up constantly when debugging, so internalize it now.


```python
def modify_list(lst):
    lst.append(100)

original = [1, 2, 3]
modify_list(original)
print(original)  # [1, 2, 3, 100] --> modified in place
```

## 12. Classes: a First Look

A class bundles data (attributes) and behavior (methods) together. `__init__` is the constructor, called automatically when you create an instance. `self` refers to the specific instance.

This is intentionally a light touch — inheritance, encapsulation, abstraction, and polymorphism get their own dedicated walkthrough in **Part 2 (OOP Fundamentals)**.


```python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    def greet(self):
        print(f"Hi, I'm {self.name} and I scored {self.score}.")

s = Student("Alice", 95)
s.greet()
```

## 13. Error Handling

`try` / `except` / `finally` lets you handle errors gracefully instead of crashing. `finally` always runs, whether or not an error occurred. Errors you'll see often this semester: `FileNotFoundError`, `KeyError`, `ValueError`, `RuntimeError` (very common with PyTorch shape mismatches), and `TypeError`.

```python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("Can't divide by zero. Returning None")
        return None
    finally:
        print("safe_divide finished")

print(safe_divide(10, 2))
print(safe_divide(5, 0))
```

## 16. NumPy Basics

This is the section to actually get comfortable with, NumPy arrays are the mental model for PyTorch tensors, and you'll be reasoning about shapes constantly in MyTorch and beyond.


```python
import numpy as np

# Creating arrays & indexing
arr = np.arange(5)
print("arr   :", arr)
print("third :", arr[2])
print("last  :", arr[-1])
```

```python
# Shape and reshaping
A = np.arange(9).reshape(3, 3)
print(A)
print("shape:", A.shape)
print("center value A[1,1]:", A[1, 1])
```

```python
# Slicing
M = np.arange(1, 10).reshape(3, 3)
print(M)

row1 = M[0, :]
col1 = M[:, 0]
block = M[:2, :2]

print("row1 :", row1)
print("col1 :", col1)
print("block:\n", block)
```

```python
# Elementwise operations & broadcasting
x = np.array([1, 2, 3])
y = np.array([10, 20, 30])

print("x + y :", x + y)
print("x ** 2:", x ** 2)
```

```python
# Elementwise (*) vs matrix multiplication (@)
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 2]])

hadamard = A * B   # elementwise product
matmul = A @ B      # matrix multiplication

print("Hadamard (A * B):\n", hadamard)
print("Matmul   (A @ B):\n", matmul)
```
